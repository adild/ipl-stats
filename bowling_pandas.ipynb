```python
import pandas as pd
```

```python
ball_df = pd.read_csv('data/IPL Ball-by-Ball 2008-2020.csv')
match_df = pd.read_csv('data/IPL Matches 2008-2020.csv')
```

```python
ball_df.info()
```

```python
pd.set_option('display.max_columns', 30)
pd.set_option('display.max_rows', 50)
```

```python
filt = ball_df['bowler'].str.contains('Agarkar', na=False)
```

```python
ball_df.loc[filt, 'bowler']
```

```python
filt = (ball_df['bowler'] == 'AB Agarkar') & \
    (ball_df['batting_team'] == 'Royal Challengers Bangalore')
bowlerAgainstTeam = ball_df.loc[filt]
bowlerAgainstTeam
```

```python
merge_result = bowlerAgainstTeam.merge(match_df, how='inner', on='id')
merge_result
```

```python
res = merge_result.groupby(
        ['id']
    ).agg(
        bowler = ('bowler', 'unique'),
        team_playing = ('bowling_team', 'unique'),
        against = ('batting_team', 'unique'),
        runs = ('total_runs', 'sum'),
        balls = ('ball', 'count'),    
        wickets = ('is_wicket', 'sum'),
        date = ('date', 'unique'),
        city = ('city', 'unique'),
        venue = ('venue', 'unique')
    )
```
