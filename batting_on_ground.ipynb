```python
import pandas as pd
```

```python
ball_df = pd.read_csv('data/IPL Ball-by-Ball 2008-2020.csv')
match_df = pd.read_csv('data/IPL Matches 2008-2020.csv')
```

```python
ball_df.info()
```

```python
match_df.info()
```

```python
pd.set_option('display.max_columns', 30)
pd.set_option('display.max_rows', 50)
```

```python
merge_result = ball_df.merge(match_df, how='inner', on='id')
```

```python
merge_result.info()
```

```python
filt = (merge_result['batsman'] == 'RG Sharma') \
& (merge_result['venue'] == 'MA Chidambaram Stadium, Chepauk')
```

```python
batsmanOnGround = merge_result.loc[filt]
```

```python
batsmanOnGround
```

```python
res = batsmanOnGround.groupby(
        ['id']
    ).agg(
        batman = ('batsman', 'unique'),
        team_playing = ('batting_team', 'unique'),
        against = ('bowling_team', 'unique'),
        runs = ('batsman_runs', 'sum'),
        balls = ('ball', 'count'),     
        date = ('date', 'unique'),
        city = ('city', 'unique'),
        venue = ('venue', 'unique')
    )
```
