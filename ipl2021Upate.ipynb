```python
import pandas as pd
```

```python
df = pd.read_csv('ipl-2021-data/all_matches.csv')
```

```python
filt = (df['striker'] == 'BB McCullum') \
& (df['bowling_team'] == 'Royal Challengers Bangalore') \
# & (df['extras'] == 0) \
# & (df['match_id'] == 392221)
```

```python
batsmanAgainstTeam = df.loc[filt]
```

```python
batsmanAgainstTeam.tail(50)
```

```python
res = batsmanAgainstTeam.groupby(
        ['match_id']
    ).agg(
        batman = ('striker', 'unique'),
        team_playing = ('batting_team', 'unique'),
        against = ('bowling_team', 'unique'),
        runs = ('runs_off_bat', 'sum'),
        balls = ('ball', 'count'),     
        date = ('start_date', 'unique'),
        venue = ('venue', 'unique')
    )
```
